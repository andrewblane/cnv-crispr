```python
array_index = pandas.read_table(open("data/arraykey_GPL9777.txt", "r"))
```

```python
cnv_ratios_and_genenames = pandas.concat([cnvratios, array_index], axis=1)
```

```python
cnv_ratios_and_genenames.iloc[0]
```

```python
cnv_ratios_and_genenames.values[0]
```

```python
# Print then remove the last line from the table; was just a footer
print(cnv_ratios_and_genenames.iloc[-1])
cnv_ratios_and_genenames = cnv_ratios_and_genenames.iloc[:-1]
```

```python
def check_ids_in_contantenated_table_match():

    for item in cnv_ratios_and_genenames.iterrows():
        if item[1]["ID_REF"] == item[1]["ID"]:
            None
        else:
            print(item[1])
            print("\n")
            print(repr(item[1]["ID_REF"]))
            print(repr(item[1]["ID"]))
            break
```

```python
check_ids_in_contantenated_table_match()
```

Hmm. They don't match; found that at some point the Pandas import decided to take the ID values as strings instead of ints. Make sure they're all ints.

```python
cnv_ratios_and_genenames["ID"] = cnv_ratios_and_genenames["ID"].apply(int)
cnv_ratios_and_genenames["ID_REF"] = cnv_ratios_and_genenames["ID_REF"].apply(int)
```

```python
check_ids_in_contantenated_table_match()
```

OK; look good now. So, we have CNV data with gene names/IDs associated.
